```python
import matplotlib.pyplot as plt
import networkx as nx

class Tugun:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

# Eski daraxt yaratish
ildiz = Tugun(4)
ildiz.left = Tugun(2)
ildiz.right = Tugun(7)
ildiz.left.right = Tugun(3)
ildiz.left.left = Tugun(1)
ildiz.right.right = Tugun(8)
ildiz.right.left = Tugun(6)

# Barglarni topish
barg = []
if ildiz.left.left is not None:
    barg.append(ildiz.left.left.value)
if ildiz.left.right is not None:
    barg.append(ildiz.left.right.value)
if ildiz.right.right is not None:
    barg.append(ildiz.right.right.value)

# Barglarni sortlash
barg.sort()

# Qayta qurish
mid = len(barg) // 2
balanced_ildiz = Tugun(barg[mid])
balanced_ildiz.left = Tugun(barg[mid - 1])
balanced_ildiz.right = Tugun(barg[mid + 1])

# Daraxtni grafik shaklida chizish funksiyasi
def chizish(ildiz, title):
    G = nx.DiGraph()
    def add_edges(node):
        if node.left:
            G.add_edge(node.value, node.left.value)
            add_edges(node.left)
        if node.right:
            G.add_edge(node.value, node.right.value)
            add_edges(node.right)
    add_edges(ildiz)
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=True, node_size=5000, node_color='skyblue', font_size=15, font_weight='bold')
    plt.title(title)
    plt.show()

# Eski daraxtni chizish
chizish(ildiz, "Eski Daraxt")

# Yangi muvozanatlangan daraxtni chizish
chizish(balanced_ildiz, "Yangi Muvozanatlangan Daraxt")
```
